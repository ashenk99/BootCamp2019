```python
def BernsteinPoly(n,v):
    p = lambda x : scipy.special.comb(n,v) * (x**v) * (1-x) **(n-v)
    return p
```

```python
class Backpack:
    """Backpack Object Class. Has name , color, and maximum size
        
    Attributes:
        name (str): name of owner
        contents (list): list of backpack contents
        color (str): backpack color
        max size (int): max number of iterms the backpack can store
    """
    def __init__(self, name, color, max_size=5):
        """Set name, initialize an empty list of contents
        
        Parameters:
            name (str): name of the backpack owner
            color(str): backpack color
            max size (int): max number of iterms the backpack can store
        """
        self.name = name
        self.contents = []
        self.color = color
        self.max_size = max_size
    def put(self, item):
        """Puts an item into the backpack's contents, if the backpack
        is full, fails to do so.
        
        Parameters:
            item (str): name of the item"""
        if len(self.contents) < 5:
            self.contents.append(item)
        else: print("No Room!")
    def take(self,item):
        """Takes an item out of the backpack. If the item isn't there,
        fails to do so.
        
        Paramters:
            item (str): name of item"""
        if item in self.contents:
            self.contents.remove(item)
    def dump(self):
        """Dumps the contents of the backpack.
        Parameters: None"""
        self.contents = []
    def __add__(self, other):
        """Add the number of contents of each Backpack"""
        return len(self.contents) + len(other.contents)
    def __eq__(self, other):
        """Returns if two backpacks are equal, that is,
        they have the same name, color and contents
        
        Parametrs: 
            other (Backpack) : backpack object for comparison
        """
        a = self.name == other.name
        b = self.color == other.color
        c = self.contents == other.contents
        if a and b and c:
            return True
        else: return False
    def __str__(self):
        """Returns a string representation of a backpack obj"""
        rep = "Owner:\t\t{}".format(self.name)
        rep += "\nColor:\t\t{}".format(self.color)
        rep += "\nSize:\t\t{}".format(len(self.contents))
        rep += "\nMax Size:\t{}".format(self.max_size)
        rep += "\nContents:\t{}".format(self.contents)
        return rep
def test_backpack():
    testpack = Backpack("Barry", "black")
    if testpack.name != "Barry":
        print("Backpack.name assigned incorrectly")
    for item in ["pencil", "pen", "paper", "computer"]:
        testpack.put(item)
    print("Contents:", testpack.contents)
        
    
```

```python
class Jetpack(Backpack):
    """Subclass of class backpack. Adds a fuel tank, and creates a fly
    method to fly the jetpack, using up fuel. Dump is overridden to dump
    bot the contents and fuel tank
    
    Attributes:
        name (str): name of owner
        contents (list): list of backpack contents
        color (str): backpack color
        max size (int): max number of iterms the jetpack can store
        fuel (float): amount of fuel in the fuel tank
    """
    def __init__(self, name, color, max_size=2, fuel=10):
        """Set name, initialize an empty list of contents

        Parameters:
            name (str): name of the backpack owner
            color(str): backpack color
            max size (int): max number of iterms the backpack can store
        """
        Backpack.__init__(self,name,color,max_size)
        self.fuel = fuel
    def fly(self, burn_amt):
        """Fly the jetpack and burn corresponding fuel, if possible
        
        Parameters:
            burn_amt (float): amount to be burned in the flight
        """
        if burn_amt > self.fuel:
            print("Not enough fuel!")
        self.fuel = self.fuel - burn_amt
    def dump(self):
        """Dumps both the jetpack contents and its fuel tank
        
        Parameters:
        None
        """
        Backpack.dump(self)
        self.fuel = 0
    
```

```python
import numpy as np
class ComplexNumber:
    """Complex Number Object Class. Has real an imaginary part
    
    Attributes:
        a (float) = real part
        b (float) = imaginary part
    """
    def __init__(self, real, imag):
        """Set real and imaginary part
        
        Parameters:
            real (float) = real part
            imag (float) = imag part
        """
        self.real = real
        self.imag = imag
    def conjugate(self):
        """Returns the complex conjugate of a complex number
        
        Parameters:
            None
        """
        x = ComplexNumber(self.real, -self.imag)
#         x.real = self.real
#         x.imag = -self.imag
        return x
    def __str__(self):
        """String representation of complex number"""
        if self.imag > 0:
            return "({}+{}j)".format(self.real, self.imag)
        
        else: return "({}{}j)".format(self.real, self.imag)
    def __abs__(self):
        """Returns the magnitude of a complex number"""
        return np.sqrt(self.real**2 + self.imag**2)
    def __eq__(self, other):
        """Returns whether or not two complex numbers are equal
        
        Params: two complex numbers
        """
        a = self.real == other.real
        b = self.imag == other.imag
        if a and b:
            return True
        else: return False
    def __add__(self, other):
        """Adds two complex numbers
        
        Params: two complex numbers
        """
        real = self.real + other.real
        imag = self.imag + other.imag
        x = ComplexNumber(real, imag)
        return x
    def __sub__(self, other):
        """Subtracts two complex numbers 
        (subtracts second from the first)
        
        Params: two complex numbers
        """
        
        real = self.real - other.real
        imag = self.imag - other.imag
        x = ComplexNumber(real, imag)
        return x
    
    def __mul__(self, other):
        """Multiplies two comple numbers
        
        Params: two complex numbers
        """
        
        real = self.real*other.real - self.imag * other.imag
        imag = self.imag*other.real + self.real * other.imag
        x = ComplexNumber(real, imag)
        return x
    def __truediv__(self, other):
        """Divides one complex (first arg) by a second"""
        print(other.conjugate())
        un_quotient = self * other.conjugate()
        print(un_quotient)
        norm = abs(other) ** 2
        print(norm)
        real = un_quotient.real / norm
        imag = un_quotient.imag / norm
        x = ComplexNumber(real, imag)
        return x
```

```python
from random import choice
def random_walk(max_iters=1e10):
    walk = 0
    directions = [1, -1]
    try :
        for i in range(int(max_iters)):
            walk += choice(directions)   
    except KeyboardInterrupt as k:
        print("Process interrupted at iteration {}".format(i))
    else: print("Process completed")
    return walk
    
```

```python
class ContentFilter:
    def __init__(self, fname):
        flag = True
        while flag:
            try:
                with open(fname, 'r') as file:
                    flag = False
                    
                    self.name = fname
                    self.contents = file.readlines()
                    self.lines = len(self.contents)
                with open(fname, 'r') as file:
                    x = file.read()
                    self.numeric = sum([s.isdigit() for s in x])
                    self.alpha = sum([s.isalpha() for s in x])
                    self.spaces = sum([s.isspace() for s in x])
                    self.totchar = sum(1 for s in x)
                    
                    
            except (FileNotFoundError, TypeError, OSError) as e:
                fname = input("Please enter a valid file name:") 
                
    def uniform(self, wname, mode='w', case = "upper"):
        if mode not in ['w', 'x', 'a']:
            raise ValueError("Invalid mode")
        contents = self.contents
        if case not in ['upper', 'lower']:
            raise ValueError("Invalid case")
        with open(wname, mode) as file:
            if case == 'upper':
                file.writelines([contents[i].upper() for i in range(len(contents))])
            else: file.write(contents.lower())
                
    def reverse(self, wname, mode='w', unit= "line"):
        if mode not in ['w', 'x', 'a']:
            raise ValueError("Invalid mode")
        if unit not in ["line", "word"]:
            raise ValueError("Invalid unit")
        contents = self.contents
        with open(wname, mode) as file:
            if unit == "line":
                for i in reversed(range(len(contents))):
                    
                    file.write(contents[i].replace("\n",""))
                    if i != 0:
                        file.write("\n")
            else:
                lines = ([line[::-1].replace("\n","")  for line in contents])
                for i in range(len(lines) -1):
                    file.write(lines[i] + "\n")
                file.write(lines[-1])
                
        
    def transpose(self, wname, mode='w'):
        if mode not in ['w', 'x', 'a']:
            raise ValueError("Invalid mode")
        with open(wname, mode) as file:
            contents = self.contents
            num_cols = len(contents)
            num_rows = len(contents[0].split())


            for i in range(num_rows):
                line= []
                for j in range(num_cols):
                    line.append(contents[j].split()[i])
                file.writelines(" ".join(line))
                if i != num_rows -1:
                    file.write('\n')
            
    def __str__(self):
        ret = ""
        ret += "Source file:\t\t\t{}".format(self.name) + '\n'
        ret += "Total characters:\t\t\t{}".format(self.totchar) + '\n'
        ret += "Alphabetic characters:\t\t\t{}".format(self.alpha) + '\n'
        ret += "Numerical characters:\t\t\t{}".format(self.numeric) + '\n'
        ret += "Whitespace characters:\t\t\t{}".format(self.spaces) + '\n'
        ret += "Number of lines:\t\t\t{}".format(self.lines)
        return ret
        
```
